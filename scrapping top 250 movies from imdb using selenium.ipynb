<h1>IMDB top 250 movies list scrapping</h1>
<h4>1. Importing Web Driver from selenium</h4>

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
```

<h4>2. Opening chrome driver with URL of the page we are scrapping</h4>

```python
driver = webdriver.Chrome()

url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"

driver.get(url)
```

<h4>3. Getting element by using class</h4>

```python
same= driver.find_elements(By.CLASS_NAME,"titleColumn")
print(len(same))
```

<h4>4. Extracting Movie Name and Year from Title</h4>

```python
names=driver.find_elements(By.CLASS_NAME,"titleColumn")
named=[]
year=[]
for i in names:
    a=i.text
    b=i.text
    index = a.find('. ')
    if index != -1:
        b= a = a[index + 2:]
    ind=a.find('(')
    if ind !=-1:
        a= a[:ind-1]
        b=int(b[ind+1:len(b)-1])
    named.append(a)
    year.append(b)
    
```

<h4>5. Getting Movie rating in Float data type</h4>

```python
rateings=driver.find_elements(By.TAG_NAME,"strong")
rating=[]
for i in rateings:
    r=i.text
    r=float(r)
    rating.append(r)
```

<h4>6. Creating DataFrame</h4>

```python
df=pd.DataFrame({'Movie Name':named,'Reliase Year':year,'Rating':rating})
```

<h4>7. Sorting list according to release date</h4>

```python
df= df.sort_values('Reliase Year', ascending=False)
print(df)
```

<h4>8. Saving as .csv file</h4>

```python
df.to_csv('complete.csv')
```
